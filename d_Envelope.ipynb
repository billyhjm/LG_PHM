## Apply the actual data

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import spectrogram
import librosa

file_path1 = './data/autorash.mp3'
y1, sr1 = librosa.load(file_path1, sr=None)

file_path2 = './data/knocking.mp3'
y2, sr2 = librosa.load(file_path2, sr=None)


a = y1 # autorash
k = y2 # knocking

print(np.shape(a))
print(np.shape(k))
print(sr1)
print(sr2)
```

```python
slice_length = len(y2) # 두 음성 신호의 길이를 맞춰주기 위한 코드

# Autorash 데이터

v_a = a[:slice_length]
T_a = len(v_a) / sr1
t_a = np.arange(1/sr1, T_a + 1/sr1, 1/sr1)

# Knocking 데이터
v_k = k
T_k = len(v_k) / sr2
t_k = np.arange(1/sr2, T_k + 1/sr2, 1/sr2)
```

```python
a_hilbert=hilbert(v_a,axis=0)
a_envelop = abs(a_hilbert)

k_hilbert=hilbert(k,axis=0)
k_envelop = abs(k_hilbert)
```

## Envelop of Autorash

```python
plt.figure(figsize=(8,12))

plt.subplot(3,1,1)
plt.plot(t_a,a_envelop,'r')
plt.xlabel('t')
plt.ylabel('x(t)')

plt.subplot(3,1,2)
plt.plot(t_a,v_a,'b')
plt.plot(t_a,a_envelop,'r')
plt.xlabel('t')
plt.ylabel('x(t)')

plt.subplot(3,1,3)
plt.plot(t_a,v_a,'b')
plt.plot(t_a,a_envelop,'r')
plt.xlim([0.60, 0.72])
plt.xlabel('t')
plt.ylabel('x(t)')

plt.subplots_adjust(hspace=0.35)
```

```python
F_original,A_original=utils.fft_simple(v_a-np.mean(v_a),sr1)

F_envelop,A_envelop=utils.fft_simple(a_envelop-np.mean(a_envelop),sr1)


plt.figure(figsize=(10,10))

plt.subplot(4,1,1)
plt.plot(F_original,A_original)
plt.xlabel('f')
plt.ylabel('|X(f)')
plt.title('Autorash single-sided FFT')
plt.xlim([0,5000])

plt.subplot(4,1,2)
plt.plot(F_original,A_original)
plt.xlabel('f')
plt.ylabel('|X(f)')
plt.title('Autorash single-sided FFT(Zoomed)')
plt.xlim([0,100])


plt.subplot(4,1,3)
plt.plot(F_envelop,A_envelop)
plt.xlabel('f')
plt.ylabel('|X(f)')
plt.title('Envelop of Autorash Single-sided FFT')
plt.xlim([0,5000])


plt.subplot(4,1,4)
plt.plot(F_envelop,A_envelop)
plt.xlabel('f')
plt.ylabel('|X(f)')
plt.title('Envelop of Autorash Single-sided FFT (Zoomed)')
plt.xlim([0,100])
plt.subplots_adjust(hspace=2)
```
